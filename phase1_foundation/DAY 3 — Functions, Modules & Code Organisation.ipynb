```python
def calculate_love_score(name1, name2):
    
    true = list('true')
    love = list('love')
    true_count = []

    for item in name1:
        if item in true:
            true_count.append(item)
            
    for item in name2:
        if item in true:
            true_count.append(item)
            
    love_count = []
    for item in name1:
        if item in love:
            love_count.append(item)
            
    for item in name2:
        if item in love:
            love_count.append(item)
            
    love_score = str(len(true_count)) + str(len(love_count))
        
    print(f'Love Score = {love_score}')


calculate_love_score("Angela Yu", "Jack Bauer")

```

```python
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
    
print(is_leap_year(2024))
```

```python
# Step 1: Create Functions for the different operators
def add(n1,n2):
    return n1 + n2

def subtract(n1, n2):
    return n1 - n2

def multiply(n1, n2):
    return n1 * n2

def divide(n1, n2):
    return n1 / n2

# Step 2: Create a dictionary called operators, with the operator's symbol as the key and its function as the value
operators = {
    "+": add,
    "-": subtract,
    "*": multiply,
    "/": divide
}

# Step 3: Ask the user for their inputs
user_input1 = int(input("Enter the first number: \n"))
user_operation = input("Enter the opertion you will like to perform:\n +\n -\n *\n /\n")
user_input2 = int(input("Enter the second number: \n"))

# Step 4: Create a function that performs all the functionality of the calculator, and then create a loop that asks the user if they
#         want to continue with the value from the previous operation, or they want to begin anew
def calculation(input1, operator, input2):
    if operator not in operators:
        print("Enter a valid operator")
        return
    else:
        result = operators[operator](input1, input2)
        print (f"{input1} {operator} {input2} = {result}")
    
        continuation = input("Would you like to continue with the previous value? (Y/N): \n").lower()
    
        while continuation == "y":
            previous_result = result
            new_input = int(input("Enter another number: \n"))
            
            result  = operators[operator](previous_result, new_input)
            print(f"{previous_result} {user_operation} {new_input} = {result}")
            
            continuation = (input("Would you like to continue with the previous value? (Y/N): \n")).lower()
            
            print(f"{previous_result} {user_operation} {new_input} = {result}")
    return result

solution = calculation(user_input1, user_operation, user_input2)
    
```
